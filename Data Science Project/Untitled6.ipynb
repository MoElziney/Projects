```python
#Add outliers:
import random
num_outliers = 13
outliers = []

#random values
platforms = ["TikTok", "Facebook", "Telegram", "YouTube"]
locations = ["North Korea", "Japan", "Germany", "Australia"]
genders = ["Female", "Male"]
demographics = ["Rural", "Urban", "Sub_Urban"]
professions = ["Unemployed", "Farmer", "Cashier", "driver"]
video_categories = ["ASMR", "Gaming", "Lifestlye"]
watch_reasons = ["Addiction", "Bad Sleeping", "Habit"]
devices = ["Windows", "Tablet", "Quantum Computer"]
oses = ["Linux", "Android", "Windows 95"]
connections = ["Satellite", "Wi-Fi", "Mobile Data"]
frequency=["Night","Afternoon","Evening"]
watch_time=["9:00 pm","5:00 pm","2:00 am"]
current_activity=["Commuting","At home","At work"]

for _ in range(num_outliers):
    outliers.append({
        "age": random.randint(80, 90),  
        "income": random.randint(22000, 25000),  
        "total time spent": random.uniform(600, 720),  
        "number of sessions": random.randint(18, 20),  
        "video length": random.uniform(60, 200),  
        "importance score": random.uniform(0, 1),  
        "time spent on video": random.uniform(50, 200),  
        "number of videos watched": random.randint(60, 70),  
        "scroll rate": random.uniform(100, 500),  
        "productivityloss": random.uniform(8, 10),  
        "satisfaction": random.uniform(0, 2), 
        "self control": random.uniform(0, 1),
        "frequency": random.choice(frequency),
        "platform": random.choice(platforms),
        "watch time": random.choice(watch_time),
        "currentactivity": random.choice(current_activity),
        "location": random.choice(locations),
        "gender": random.choice(genders),
        "demographics": random.choice(demographics),
        "profession": random.choice(professions),
        "owns property": random.choice([True, False]),
        "indebt": random.choice([True, False]),
        "video category": random.choice(video_categories),
        "watch reason": random.choice(watch_reasons),
        "devicetype": random.choice(devices),
        "os": random.choice(oses),
        "connectiontype": random.choice(connections)
    })
outliers_df = pd.DataFrame(outliers)
merged_df = pd.concat([merged_df, outliers_df], ignore_index=True)

merged_df
```

```python
#question 1:
selected=["platform","location","satisfaction","indebt","owns property"]
copy_df=merged_df[selected]
copy_df
```

```python
filtered_df = copy_df[(copy_df["location"] == "United States") & (~copy_df["indebt"]) & (copy_df["owns property"])]

platform_means = filtered_df.groupby("platform")["satisfaction"].mean()

platform_means.plot(kind="bar")

plt.xlabel("Platform")
plt.ylabel("Mean Satisfaction")
plt.title("Satisfaction for Platforms in United States for normal people")
plt.xticks(rotation=45)  

plt.show()
```

```python
labels = ["Indebt without home", "Owns Property without dept","have home and indept"," No home No dept"]
counts = [
    copy_df[(copy_df["indebt"] == True) & (copy_df["owns property"] == False)].shape[0],
    copy_df[(copy_df["indebt"] == False) & (copy_df["owns property"] == True)].shape[0],
    copy_df[(copy_df["indebt"] == True) & (copy_df["owns property"] == True)].shape[0],
    copy_df[(copy_df["indebt"] == False) & (copy_df["owns property"] == False)].shape[0]
]

plt.figure(figsize=(6, 6))
plt.pie(counts, labels=labels, autopct='%1.1f%%', colors=["red", "blue","green","yellow"], startangle=90)
plt.title("Comparison of Indebt  vs Owns Property ")
plt.show()
```

```python
selected=["income","indebt","owns property","video category"]
copy_df5=merged_df[selected]
copy_df5
```

```python

income_pivot = copy_df5.pivot_table(index="video category", columns="indebt", values="income", aggfunc="mean")


plt.figure(figsize=(10, 6))
sns.heatmap(income_pivot, annot=True,  fmt=".0f",cmap="coolwarm", linewidths=0.5)
plt.title("Average Income per Video Category and Financial Status")
plt.xlabel("Financial Status")
plt.ylabel("Video Category")
plt.show()
```

```python
copy_df5["indebt"] = copy_df5.apply(lambda row: 
                                  "High Worth" if row["owns property"] and not row["indebt"] else 
                                  "Middle Class" if row["owns property"] and row["indebt"] else
                                  "Struggling" if not row["owns property"] and row["indebt"] else
                                  "Emerging", axis=1)

financial_video_counts = copy_df5.groupby(["video category", "indebt"]).size().unstack()

# رسم المخطط العمودي
financial_video_counts.plot(kind="bar", stacked=True, figsize=(12, 6), colormap="viridis")
plt.title("Video Content Consumption by Financial Status")
plt.xlabel("Video Category")
plt.ylabel("Number of Viewers")
plt.xticks(rotation=45)
plt.legend(title="Financial Status")
plt.show()
```
